```python
#loading required libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Reading nobel prize data
nobel=pd.read_csv(r'E:\Data science\Projects and datasets\nobel.csv')

#taking a look at first several winners
nobel.head(n=6)
```

```python
#Display number of male and female winners
display(nobel['sex'].value_counts())

#Display numbers of country winners
nobel['birth_country'].value_counts().head(n=10)
```

```python
#Calculating the proportion of USA born winners per decade
nobel['usa_born_winner'] = nobel['birth_country']=='United States of America' 
nobel['decade']=round(np.round(nobel['year']/10)*10).astype(int)
nobel.groupby('decade')['usa_born_winner'].mean()
```

```python
# Setting the plotting theme
sns.set()
# and setting the size of all plots.
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [11, 7]
# Plotting USA born winners 
ax = sns.lineplot(data=nobel,x='decade',y='usa_born_winner')

# Adding %-formatting to the y-axis
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter())
```

```python
# calculating number of female winners
nobel['female_winner']=nobel['sex']=='Female'
prop_female_winner=nobel.groupby(['decade','category'],as_index=False)['female_winner'].mean()

#plotting Female winner
ax=sns.lineplot(data=prop_female_winner,x='decade',y='female_winner',hue='category')
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter())
```

```python
#finding how old they were when they got the prize
nobel['birth_date']=pd.to_datetime(nobel['birth_date'])
nobel['age']=nobel['year']-nobel['birth_date'].dt.year
nobel
#visualizing age nobel prize winners
sns.lmplot(data=nobel,x='year',y='age')
```

```python
#visualize Age differences between prize categories
sns.lmplot(x='year',y='age',row='category',data=nobel)
```

```python
# show the oldest and the youngest winners
display(nobel.nlargest(n=1,columns='age'))
nobel.nsmallest(n=1,columns='age')
```
